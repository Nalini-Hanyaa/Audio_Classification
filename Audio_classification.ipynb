```python
import librosa
import numpy as np

def analyze_rhythm(audio_path):
    # Load audio and extract tempo and beats
    audio, sr = librosa.load(audio_path)
    tempo, beats = librosa.beat.beat_track(y=audio, sr=sr)
    
    # Compute inter-beat intervals
    beat_times = librosa.frames_to_time(beats, sr=sr)
    # print(beat_times)
    beat_intervals = np.diff(beat_times)
    
    return tempo, beat_intervals

def compare_rhythm(audio_path1, audio_path2):
    # Analyze rhythm for both audio clips
    tempo1, beat_intervals1 = analyze_rhythm(audio_path1)
    tempo2, beat_intervals2 = analyze_rhythm(audio_path2)
    
    # Check if the tempos are close
    tempo_difference = abs(tempo1 - tempo2)
    if tempo_difference < 5:  # Adjust this threshold as needed
        return "Rhythm of 2 songs is same"
    
    # Align the beat intervals arrays
    min_length = min(len(beat_intervals1), len(beat_intervals2))
    beat_intervals1 = beat_intervals1[:min_length]
    beat_intervals2 = beat_intervals2[:min_length]
    
    # Compute the distance between beat intervals
    distance = np.linalg.norm(beat_intervals1 - beat_intervals2)
    
    # Determine if the rhythms are the same or different based on the distance
    if distance < 1:  # Adjust this threshold as needed
        return "Rhythm of 2 songs is same"
    else:
        return "Rhythm of 2 songs is different"

# Example usage
audio_path1 = r"D:\flute.mp3"
audio_path2 = r"D:\flute_tune.mp3"
result = compare_rhythm(audio_path1, audio_path2)
print(result)
```

```python
# import librosa
# import numpy as np

# def extract_spectral_features(audio_path):
#     # Load audio and extract spectral features
#     audio, sr = librosa.load(audio_path)
#     spectral_features = librosa.feature.chroma_stft(y=audio, sr=sr)
    
#     return spectral_features

# def normalize_feature_length(feature1, feature2):
#     # Normalize the length of the two feature arrays by padding with zeros
#     max_length = max(feature1.shape[1], feature2.shape[1])
#     padded_feature1 = np.pad(feature1, ((0, 0), (0, max_length - feature1.shape[1])), mode='constant')
#     padded_feature2 = np.pad(feature2, ((0, 0), (0, max_length - feature2.shape[1])), mode='constant')
    
#     return padded_feature1, padded_feature2

# def compare_instruments(audio_path1, audio_path2):
#     # Extract spectral features for both audio clips
#     features1 = extract_spectral_features(audio_path1)
#     features2 = extract_spectral_features(audio_path2)
    
#     # Normalize the length of the feature arrays
#     features1, features2 = normalize_feature_length(features1, features2)
    
#     # Compute similarity metric (e.g., cosine similarity) between spectral features
#     similarity = np.dot(features1.flatten(), features2.flatten()) / (np.linalg.norm(features1) * np.linalg.norm(features2))
    
#     # Determine if the instruments are the same or different based on the similarity
#     if similarity > 0.7:  # Adjust this threshold as needed
#         return "Instruments in the two audio clips are similar"
#     else:
#         return "Instruments in the two audio clips are different"

# # Example usage
# audio_path1 = r"D:\krishna_flute.mp3"
# audio_path2 = r"D:\flute.mp3"
# result = compare_instruments(audio_path1, audio_path2)
# print(result)
import librosa
import numpy as np

def extract_spectral_features(audio_path):
    # Load audio and extract spectral features
    audio, sr = librosa.load(audio_path)
    stft = np.abs(librosa.stft(audio))
    mfccs = librosa.feature.mfcc(y=audio, sr=sr)
    spectral_features = np.concatenate((stft.flatten(), mfccs.flatten()))
    return spectral_features

def compare_instruments(audio_path1, audio_path2):
    # Extract spectral features for both audio clips
    features1 = extract_spectral_features(audio_path1)
    features2 = extract_spectral_features(audio_path2)
    
    # Truncate the longer feature vector to match the length of the shorter one
    min_length = min(len(features1), len(features2))
    features1 = features1[:min_length]
    features2 = features2[:min_length]
    
    # Compute similarity metric (e.g., cosine similarity) between spectral features
    similarity = np.dot(features1, features2) / (np.linalg.norm(features1) * np.linalg.norm(features2))
    
    # Determine if the instruments are the same or different based on the similarity
    if similarity > 0.7:  # Adjust this threshold as needed
        return "Instruments in the two audio clips are similar"
    else:
        return "Instruments in the two audio clips are different"

# Example usage
audio_path1 = r"D:\flute_tune.mp3"
audio_path2 = r"D:\flute.mp3"
result = compare_instruments(audio_path1, audio_path2)
print(result)
```
